```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1234, stratify=y, test_size=.30)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
log_reg = LogisticRegression()
knn = KNeighborsClassifier()
trees = DecisionTreeClassifier()
rand_forest = RandomForestClassifier()
svm = SVC()
naive = GaussianNB()
xgb = GradientBoostingClassifier()
```

```python
# TUNING HYPERPARAMETERS
grid_svm = [{'kernel': ['rbf'],'C': [0.1, 0.5, 1]},
            {'kernel': ['linear', 'poly'],  'C': [0.1, 0.5, 1]}]

grid_rand_forest = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100, 110],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10],
    'n_estimators': [100, 200, 300, 500]}

#grid_log_reg = {'C': [0.01, 0.1, 1]}

grid_knn = {'n_neighbors':[2,3,4,5,6,7,8,9,10,12,15,18,20,25], 'metric':['manhattan','euclidean']}

grid_tree = {'criterion':['gini','entropy'],'max_depth':[3,4,5,6,7,8,9,10,20,50,75,100]}

grid_xgb = {"criterion": ["friedman_mse",  "mae"],
              "loss":["deviance","exponential"],
              "max_features":["log2","sqrt"],
              'learning_rate': [0.01,0.05,0.1,1],
              'max_depth': [3,4,5],
              'min_samples_leaf': [4,5,6],
              'subsample': [0.6,0.7,0.8],
              'n_estimators': [50,100]}



#log_reg_cv = GridSearchCV(log_reg, grid_log_reg, cv=5, return_train_score=True)
#log_reg_cv.fit(X_train,y_train)
#pred_log_reg_cv = log_reg_cv.predict(X_test)
#score_log_reg_cv = log_reg_cv.score(X_test,y_test)


knn_cv = GridSearchCV(knn, grid_knn, cv=5, return_train_score=True)
knn_cv.fit(X_train,y_train)
pred_knn_cv = knn_cv.predict(X_test)
score_knn_cv = knn_cv.score(X_test,y_test)


svm_cv = GridSearchCV(svm, grid_svm, cv=5, return_train_score=True)
svm_cv.fit(X_train,y_train)
pred_svm_cv = svm_cv.predict(X_test)
score_svm_cv = svm_cv.score(X_test,y_test)

tree_cv = GridSearchCV(trees, grid_tree, cv=5, return_train_score=True)
tree_cv.fit(X_train,y_train)
pred_tree_cv = tree_cv.predict(X_test)
score_tree_cv = tree_cv.score(X_test,y_test)


rand_forest_cv = GridSearchCV(rand_forest, grid_rand_forest, cv=5, return_train_score=True)
rand_forest_cv.fit(X_train, y_train)
pred_rand_forest_cv = rand_forest_cv.predict(X_test)
score_rand_forest_cv = rand_forest_cv.score(X_test, y_test)


xgb_cv = GridSearchCV(xgb, grid_xgb, cv=5, return_train_score=True)
xgb_cv.fit(X_train, y_train)
pred_xgb_cv = xgb_cv.predict(X_test)
score_xgb_cv = xgb_cv.score(X_test, y_test)
```

```python
# ACCURACY
accuracy = [score_knn_cv, score_tree_cv, score_rand_forest_cv, score_svm_cv, score_xgb_cv] # manca la logistica
algorithm = ['KNN', 'DECISION TREE', 'RANDOM FOREST', 'SVM', 'XGB']
data_plot = pd.DataFrame(accuracy,algorithm)
sns.catplot(data=data_plot, x=accuracy, y=algorithm, kind='bar', col_order=accuracy.sort(),height=6,aspect=2)
```
